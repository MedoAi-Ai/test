####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>
</details>


````python
import matplotlib.pyplot as plt

# Ensure 'Year' column is int and sales are numeric
df['Year'] = df['Year'].astype(int)
df['Automobile_Sales'] = pd.to_numeric(df['Automobile_Sales'], errors='coerce')

# Filter only necessary columns
df_filtered = df[['Year', 'Vehicle_Type', 'Recession', 'Automobile_Sales']]

# Group by Year and Vehicle Type, sum sales
grouped = df_filtered.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().reset_index()

# Pivot for plotting
pivot_df = grouped.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Plotting
plt.figure(figsize=(12, 7))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='o', label=column)

# Highlight recession periods
recession_years = df[df['Recession'] == 1]['Year'].unique()
for year in recession_years:
    plt.axvline(x=year, color='gray', linestyle='--', alpha=0.3)

# Labels & legend
plt.title('Automobile Sales Trend by Vehicle Type (with Recession Periods)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.tight_layout()
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure Year and GDP are in correct format
df['Year'] = df['Year'].astype(int)
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')

# Separate data for recession and non-recession periods
gdp_recession = df[df['Recession'] == 1].groupby('Year')['GDP'].mean().reset_index()
gdp_non_recession = df[df['Recession'] == 0].groupby('Year')['GDP'].mean().reset_index()

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

# Plot GDP during recession
axes[0].plot(gdp_recession['Year'], gdp_recession['GDP'], marker='o', color='red')
axes[0].set_title('GDP During Recession')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('GDP')
axes[0].grid(True)

# Plot GDP during non-recession
axes[1].plot(gdp_non_recession['Year'], gdp_non_recession['GDP'], marker='o', color='green')
axes[1].set_title('GDP During Non-Recession')
axes[1].set_xlabel('Year')
axes[1].grid(True)

# Overall title and layout
plt.suptitle('Comparison of GDP During Recession and Non-Recession Periods')
plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure necessary columns are numeric
df['Automobile_Sales'] = pd.to_numeric(df['Automobile_Sales'], errors='coerce')
df['Seasonality_Weight'] = pd.to_numeric(df['Seasonality_Weight'], errors='coerce')
df['Month'] = pd.Categorical(df['Month'], categories=[
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
], ordered=True)

# Group by Month to get average sales and seasonality weight
seasonality_data = df.groupby('Month')[['Automobile_Sales', 'Seasonality_Weight']].mean().reset_index()

# Plot bubble chart
plt.figure(figsize=(12, 6))
plt.scatter(
    seasonality_data['Month'],
    seasonality_data['Automobile_Sales'],
    s=seasonality_data['Seasonality_Weight'] * 1000,  # scale bubbles
    alpha=0.6,
    color='teal',
    edgecolors='black'
)

# Label plot
plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Month')
plt.ylabel('Average Automobile Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

 ### TASK 1.9: Develop a lineplot to analyse the effect of the unemployment rate on vehicle type and sales during the Recession Period.

 #### You can create a lineplot and title the plot as 'Effect of Unemployment Rate on Vehicle Type and Sales'


<details><summary>Click here for a hint</summary>

    Filter out the data for recession period<br>
     Make use of lineplot() from seaborn and pass the relavent data</p>

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for recession period
recession_data = df[df['Recession'] == 1]

# Group by Year, Vehicle Type, and calculate average sales and unemployment rate
grouped_data = recession_data.groupby(['Year', 'Vehicle_Type']).agg({
    'Automobile_Sales': 'sum',
    'unemployment_rate': 'mean'
}).reset_index()

# Set up the plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped_data, x='unemployment_rate', y='Automobile_Sales', hue='Vehicle_Type', marker='o')

# Customize
plt.title('Effect of Unemployment Rate on Vehicle Sales by Type During Recession')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Total Automobile Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
````
